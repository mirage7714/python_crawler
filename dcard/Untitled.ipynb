```python
import requests
import json
from bs4 import BeautifulSoup as soup
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
```

```python
dcard_url = 'https://www.dcard.tw/f/softwareengineer?latest=true'
driver = webdriver.Chrome()
driver.get(dcard_url)
```

```python
request = requests.get(dcard_url)
print(request.status_code)
```

```python
content = soup(request.text, 'lxml')
```

```python
divs = content.find_all('a', href=lambda href: href and "/f/softwareengineer/p" in href)
```

```python
for div in divs:
    print(div['href'])
```

```python
def getPageContent():
    content= soup(driver.page_source, 'lxml')
    links = content.find_all('a', href=lambda href: href and "/f/softwareengineer/p" in href)
    return links
```

```python
def getPostEntry():
    content = soup(driver.page_source, 'lxml')
    
    post_entries = content.find_all("//div[contains(@class, 'PostList_entry')]")
    return post_entries
```

```python
for n in range(10):
    n += 1
    #links = getPageContent()
    posts = getPostEntry()
    for link in posts:
        print(link.text)
    driver.execute_script('window.scrollTo({}, {})'.format(0+n*1000, (n+1)*1000))
    
```
