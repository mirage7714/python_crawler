```python
import requests
from bs4 import BeautifulSoup as soup
import random
import time
```

為了讓程式比較不容易被防爬蟲程式阻擋，這邊使用隨機選出的user-agent

```python
def GetUserAgent():
    agent_list = []
    with open('useragent_list.txt','r+') as f:
        agents = f.readlines()
    for agent in agents:
        agent_list.append(agent.replace('\n',''))
    index = int(random.random()* len(agent_list))
    headers = {'User-Agent':agent_list[index]}
    return headers
```

抓出這個看板一共有多少頁數，並回傳總頁數

```python
def GetNextPage(content):
    next_page = content.find('div',{'class':'btn-group btn-group-paging'}).find_all('a',{'class':'btn wide'})
    next_page_url = next_page[1].attrs['href']
    page_number = next_page_url.split('index')[1].split('.')[0]
    return page_number
```

```python
def GetUrl(content):
    page_url = content.find('a').attrs['href']
    return page_url
```

抓出頁面中的所有文章連結

```python
def GetPageLinks(url, main_url, headers):
    page_req = requests.get(url, headers = headers)
    cont = soup(page_req.text, 'lxml')
    links = cont.find_all('div',{'class':'r-ent'})
    link_list = []
    for link in links:
        try:
            urls = link.find_all('a')
            for u in urls:
                link_list.append(main_url+u.attrs['href'])
        except:
            continue
    return link_list
```

抓出文章中的標題、作者、發文時間、內文、以及推文內容，加入一個dict後回傳

```python
def GetPageContent(url, headers):
    po_cont = {}
    req = requests.get(url, headers = headers)
    tree = soup(req.text, 'lxml')
    meta = tree.find_all('span',{'class':'article-meta-value'})
    cont = req.text.split()
    pushes = tree.find_all('span',{'class':'f3 push-content'})
    push_list = []
    for push in pushes:
        push_list.append(push.text.replace(': ',''))
    po_cont['push'] = push_list
    po_cont['author'] = meta[0].text
    po_cont['title'] = meta[2].text
    po_cont['time'] = meta[3].text
    cont = req.text.split(meta[3].text)[1].split('--')[0].replace('\n',' ').replace('</span></div>','')
    po_cont['cont'] = cont
    return po_cont
```

```python
board = 'GirlsFront'
url = 'https://www.ptt.cc'
first_page = url+'/bbs/'+board+'/index.html'
results = requests.get(first_page, GetUserAgent())
content = soup(results.text, 'lxml')
links = content.find_all('div',{'class':'r-ent'})
page_number= GetNextPage(content)
file= board+'.txt'
```

```python
with open(file, 'a+') as f:
    for link in links:
        try:
            each_page_url = url+GetUrl(link)
            headers = GetUserAgent()
            po_cont = GetPageContent(each_page_url, headers)
            f.write(str(po_cont)+'\n')
            time.sleep(random.random() * 5)
        except:
            continue
    f.flush()
    for num in range(int(page_number),1,-1):
        page_url = url+'/bbs/'+board+'/index'+str(num)+'.html'
        page_list = GetPageLinks(page_url, url, GetUserAgent())
        print(page_url)
        for page in page_list:
            try:
                detail = GetPageContent(page, GetUserAgent())
                f.write(str(detail)+'\n')
            except:
                continue
        f.flush()
        time.sleep(random.random() * 5)
    f.flush()
```
